```python
fig = px.scatter(df, x="Anime Name", y="Number of Episodes", 
                 color="Genre", 
                 title="üì∫ Number of Episodes per Anime (by Genre)",
                 hover_data=["Anime Name", "Ratings", "Budget (in Million USD)"],  
                 template="plotly_dark",
                 size_max=10)  

fig.update_layout(
    xaxis_title="Anime Name",
    yaxis_title="Number of Episodes",
    font=dict(size=13),
    xaxis_tickangle=-45,  
    yaxis=dict(showgrid=True, gridcolor="lightgrey"),
    xaxis=dict(showgrid=True, gridcolor="lightgrey"),
    height=600, width=900  
)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
fig = px.bar(df, x="Number of Episodes", y="Anime Name", 
             title="üì∫ Number of Episodes per Anime",
             text="Number of Episodes", 
             color="Number of Episodes",  
             orientation="h",  
             color_continuous_scale="blues")  

fig.update_traces(texttemplate='%{text}', textposition="outside")

fig.update_layout(
    xaxis_title="Number of Episodes",
    yaxis_title="Anime Name",
    font=dict(size=13),
    template="plotly_dark", 
    yaxis=dict(tickmode="linear"), 
)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
fig = px.scatter_matrix(df, 
                        dimensions=['Ratings', 'Number of Episodes', 'Budget (in Million USD)', 'Release Year'],
                        color='Ratings',  
                        title="üìä Pairplot of Anime Dataset",
                        template="plotly_dark")

fig.update_layout(
    font=dict(size=12),
    width=900, height=900
)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
fig = px.density_contour(df, x="Ratings", y="Budget (in Million USD)", 
                         title="üìà KDE Density Plot: Ratings vs Budget")

fig.update_layout(
    xaxis_title="Ratings",
    yaxis_title="Budget (Million USD)",
    template="plotly_dark"
)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
df1 = df[df['Budget (in Million USD)'] > 130].sort_values(by='Budget (in Million USD)', ascending=False).head(5)

fig = px.bar(df1, x="Anime Name", y="Budget (in Million USD)", 
             title="üí∞ Most High-Budget Anime",
             text="Budget (in Million USD)",  
             color="Anime Name",  
             color_discrete_sequence=px.colors.qualitative.Set1)  

fig.update_traces(texttemplate='%{text:.1f}M', textposition="outside")

fig.update_layout(
    xaxis_title="Anime Name",
    yaxis_title="Budget (Million USD)",
    font=dict(size=13),
    template="plotly_dark", 
    xaxis_tickangle=-15  
)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
fig = px.histogram(df, x="Ratings", nbins=15, marginal="box", 
                   color_discrete_sequence=["royalblue"], opacity=0.8)

fig.update_layout(
    title="üìä Distribution of Anime Ratings",
    xaxis_title="Ratings",
    yaxis_title="Count",
    template="plotly_dark",  
    font=dict(size=13),
    bargap=0.1,  
)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
fig = px.box(df, x="Genre", y="Budget (in Million USD)", 
             color="Genre",  
             template="plotly_dark", 
             boxmode="group")

fig.update_layout(
    title="üí∞ Anime Budget Distribution by Genre",
    xaxis_title="Genre",
    yaxis_title="Budget (Million USD)",
    font=dict(size=13),
    xaxis_tickangle=-45,  
    showlegend=False  
)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
fig = px.scatter(df, x="Budget (in Million USD)", y="Ratings", 
                 color="Genre",  
                 hover_data=["Anime Name", "Animation Studio Name"],  
                 template="plotly_dark",  
                 opacity=0.8)  

fig.update_layout(
    title="üí∞ Budget vs Ratings of Anime",
    xaxis_title="Budget (Million USD)",
    yaxis_title="Ratings",
    font=dict(size=13),
    legend_title="Genre"
)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
top_studios = df['Animation Studio Name'].value_counts().head(10).reset_index()
top_studios.columns = ['Studio', 'Count']

fig = px.bar(top_studios, x='Studio', y='Count', 
             text='Count', color='Count',
             color_continuous_scale='Blues',
             title="üé¨ Top 10 Animation Studios by Number of Animes")

fig.update_traces(textposition='outside')
fig.update_layout(
    template="plotly_dark",
    font=dict(size=12, color="white"),
    title_font_size=20)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
import plotly.figure_factory as ff

corr_matrix = df.corr(numeric_only=True).round(2)
heatmap = ff.create_annotated_heatmap(
    z=corr_matrix.values, 
    x=list(corr_matrix.columns), 
    y=list(corr_matrix.index), 
    colorscale='Cividis',  
    showscale=True)

heatmap.update_layout(
    title="üî• Correlation Matrix of Anime Dataset",
    font=dict(size=12, color="white"),
    template="plotly_dark",
    title_font_size=20)

heatmap.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
fig = px.pie(df, names="Genre", 
             title="üé≠ Genre Distribution of Anime",
             color_discrete_sequence=px.colors.qualitative.Set2, 
             hole=0.3)  

fig.update_traces(textinfo="percent+label", 
                  pull=[0.05 if i == df['Genre'].value_counts().idxmax() else 0 for i in df['Genre'].unique()],  # Pops out the most common genre
                  marker=dict(line=dict(color='black', width=1)))  

fig.update_layout(
    showlegend=True,
    font=dict(size=13),
    template="plotly_dark"  
)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
fig = px.treemap(df, path=['Genre'], values='Ratings', 
                 color='Ratings', color_continuous_scale='viridis',
                 title="üåü Popularity of Anime Genres")

fig.update_layout(margin=dict(t=50, l=25, r=25, b=25),template="plotly_dark")
fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
fig = px.scatter(df, x="Release Year", y="Ratings", 
                 animation_frame="Release Year", 
                 size="Budget (in Million USD)", 
                 color="Genre",
                 hover_name="Anime Name",
                 title="üìà Anime Ratings Over Time",
                 color_continuous_scale="Plasma")

fig.update_layout(
    template="plotly_dark",  
    font=dict(size=12, color="white"), 
    plot_bgcolor="#222", 
    paper_bgcolor="#222", 
    title_font_size=20)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```

```python
fig = px.choropleth(df, locations="Most Watched in Country", 
                    locationmode='country names', 
                    color="Ratings",
                    hover_name="Most Watched in Country",
                    title="üåç Most Watched Anime by Country",
                    color_continuous_scale="tealrose")

fig.update_layout(
    geo=dict(showcoastlines=True, showland=True, landcolor="lightgray"),
    template="plotly_dark",
    font=dict(size=12, color="white"),
    title_font_size=20)

fig.show(renderer = "iframe")
fig.write_image("plotly_chart.png")
```
